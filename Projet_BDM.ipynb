````python
#classe pour l'ACP
from sklearn.decomposition import PCA
#instanciation
acp = PCA(svd_solver='full')
````

````python
#affichage des paramÃ¨tres
print(acp)
````

````python
#variance expliquÃ©e
print(acp.explained_variance_)
````

````python
#valeur corrigÃ©e
n=1967225  #21
eigval = (n-1)/n*acp.explained_variance_
print(eigval)
````

````python
#ou bien en passant par les valeurs singuliÃ¨res
print(acp.singular_values_**2/n)
````

````python
#scree plot
import matplotlib.pyplot as plt
p=21
plt.plot(numpy.arange(1,p+1),eigval)
plt.title("Scree plot")
plt.ylabel("Eigen values")
plt.xlabel("Factor number")
plt.show()
````

````python
#cumul de variance expliquÃ©e
plt.plot(numpy.arange(1,p+1),numpy.cumsum(acp.explained_variance_ratio_))
plt.title("Explained variance vs. # of factors")
plt.ylabel("Cumsum explained variance ratio")
plt.xlabel("Factor number")
plt.show()
````

````python
#seuils pour test des bÃ¢tons brisÃ©s
bs = 1/numpy.arange(p,0,-1)
bs = numpy.cumsum(bs)
bs = bs[::-1]
````

````python
import numpy as np
def plot_2d_space(X, y, label='Classes'):   
    colors = ['#1F77B4', '#FF7F0E']
    markers = ['o', 's']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==l, 0],
            X[y==l, 1],
            c=c, label=l, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()
````

````python
from sklearn.decomposition import PCA

y=fraude_train['FlAgImpAye']
pca = PCA(n_components=2)
X = pca.fit_transform(fraude_train_act)

plot_2d_space(X, y, 'Imbalanced dataset (2 PCA components)')
````

````python
from imblearn.under_sampling import TomekLinks
tl = TomekLinks() #return_indices=True, ratio='majority'
X_tl, y_tl = tl.fit_sample(X, y)

#print('Removed indexes:', id_tl)

plot_2d_space(X_tl, y_tl, 'Tomek links under-sampling')
````

````python
fraude_test_act = fraude_test.drop(['ZIBZIN','IDAvisAutorisAtionCheque','DAteTrAnsAction','FlAgImpAye'],axis='columns')
````

````python
ytrain_keras = fraude_train.iloc[0:1677541,:]
ytrain_keras = np.array(ytrain_keras['FlAgImpAye'], dtype="uint8")
ytest_keras = np.array(fraude_test['FlAgImpAye'], dtype="uint8")
yvalid_keras = fraude_train.iloc[1677541:1967225,:]
yvalid_keras = np.array(yvalid_keras['FlAgImpAye'], dtype="uint8")
#1967225
````

````python
xvalid_keras= fraude_train_act.iloc[1677541:1967225,:]
xtrain_keras = fraude_train_act.iloc[0:1677541,:]
xtest_keras = fraude_test_act
````

````python
counts = np.bincount(ytrain_keras)
print(
    "Number of positive samples in training data: {} ({:.2f}% of total)".format(
        counts[1], 100 * float(counts[1]) / len(ytrain_keras)
    )
)

weight_for_0 = 1.0 / counts[0]
weight_for_1 = 1.0 / counts[1]
````

````python
#Normalisation des donÃ©nes 


mean = np.mean(xtrain_keras, axis=0)
xtrain_keras -= mean
xvalid_keras -= mean
std = np.std(xtrain_keras, axis=0)
xtrain_keras /= std
xvalid_keras /= std
````

````python
from tensorflow import keras

model = keras.Sequential(
    [
        keras.layers.Dense(
            256, activation="relu", input_shape=(xtrain_keras.shape[-1],)
        ),
        keras.layers.Dense(256, activation="relu"),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(256, activation="relu"),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(1, activation="sigmoid"),
    ]
)
model.summary()
````

````python
metrics = [
    keras.metrics.FalseNegatives(name="fn"),
    keras.metrics.FalsePositives(name="fp"),
    keras.metrics.TrueNegatives(name="tn"),
    keras.metrics.TruePositives(name="tp"),
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
]

model.compile(
    optimizer=keras.optimizers.Adam(1e-2), loss="binary_crossentropy", metrics=metrics
)

callbacks = [keras.callbacks.ModelCheckpoint("fraud_model_at_epoch_{epoch}.h5")]
class_weight = {0: weight_for_0, 1: weight_for_1}

model.fit(
    xtrain_keras,
    ytrain_keras,
    batch_size=2048,
    epochs=30,
    verbose=2,
    callbacks=callbacks,
    validation_data=(xvalid_keras, yvalid_keras),
    class_weight=class_weight,
)
````

````python
#activation functions to try out 
activations = ['relu','sigmoid','tanh']

#dictionnary to store results 
activation_results={}
for funct in activations:
    model = get_model(act_function=funct)
    history = model.fit(X_train,y_train,
                       valdiation_data = (X_test, y_test),
                       epochs=100, verbose=0)
    activation_results[funct]=history
````
